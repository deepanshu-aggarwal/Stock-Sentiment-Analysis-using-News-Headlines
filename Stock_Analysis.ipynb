```python
# Removing Punctuations
train_data=train.iloc[:,2:27]
train_data.replace('[^a-zA-Z]',' ',regex=True,inplace=True)  
test_data=test.iloc[:,2:27]
test_data.replace('[^a-zA-Z]',' ',regex=True,inplace=True) 
```

```python
# Renaming column names for ease of access
list1=[i for i in range(25)]
new_index=[str(i) for i in list1]
train_data.columns=new_index
test_data.columns=new_index
```

```python
# Converting headlines to lower case
for index in new_index:
    train_data[index]=train_data[index].str.lower()
for index in new_index:
    test_data[index]=test_data[index].str.lower()  
```

```python
from nltk.stem import PorterStemmer
import re
import bs4
def preprocess(x):
    x = str(x).lower()
    x = x.replace(",000,000", "m").replace(",000", "k").replace("′", "'").replace("’", "'")\
                           .replace("won't", "will not").replace("cannot", "can not").replace("can't", "can not")\
                           .replace("n't", " not").replace("what's", "what is").replace("it's", "it is")\
                           .replace("'ve", " have").replace("i'm", "i am").replace("'re", " are")\
                           .replace("he's", "he is").replace("she's", "she is").replace("'s", " own")\
                           .replace("%", " percent ").replace("₹", " rupee ").replace("$", " dollar ")\
                           .replace("€", " euro ").replace("'ll", " will")
    x = re.sub(r"([0-9]+)000000", r"\1m", x)
    x = re.sub(r"([0-9]+)000", r"\1k", x)
    
    
    porter = PorterStemmer()
    pattern = re.compile('\W')
    
    if type(x) == type(''):
        x = re.sub(pattern, ' ', x)
    
    
    if type(x) == type(''):
        x = porter.stem(x)
        example1 = bs4.BeautifulSoup(x)
        x = example1.get_text()
               
    
    return x
```

```python
for i in range(25):
  train_data.iloc[:,i]=train_data.iloc[:,i].fillna('').apply(preprocess)
```

```python
for i in range(25):
  test_data.iloc[:,i]=test_data.iloc[:,i].fillna('').apply(preprocess)
```
